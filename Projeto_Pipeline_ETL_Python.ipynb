Passo 3 – Transformação (usar Cohere para enriquecer os dados)

```python
#Instalar IA Generativa Cohere
"""Eu instalei a Cohehe como exemplo de IA, pois tinha créditos para a API, poderia ser qualquer outra
,por exemplo: A API da Open AI)"""
#Instalar IA Generativa Cohere
!pip install cohere
```

```python
"""Aqui eu executo o modelo de IA 'command-a-03-2025' da Cohere, testei com outros modelos também, e sempre
mostra no resultado as frases geradas por IA de forma abreviada ,só poderá ver as frases completas se abrir
os arquivos 'csv' e 'json' gerados no passo 4: Carregamento."""
#Importar IA cohere
import cohere

# Inicializar cliente Cohere (substitua pela sua chave)
co = cohere.Client("Mrk4ABBk1D5ZVi7LWgAlhPf8I0tXRBI1hdbghf6L")

# Função para gerar mensagem personalizada
def gerar_mensagem(nome, interesses):
    prompt = f"""
    Você é um assistente de educação financeira.
    Crie uma mensagem personalizada para {nome}, considerando seus interesses: {interesses}.
    A mensagem deve ser curta e objetiva e ter pelo menos 3 frases sem cortar nada no texto principal, incluir dicas práticas e ser amigável.
    """

    response = co.chat(
        model="command-a-03-2025",  # modelo de geração mais robusto
        message=prompt,
        temperature=0.7,
        max_tokens=60
    )
    return response.text.strip()

# Criar nova coluna com mensagens
df["Mensagem_Personalizada"] = df.apply(
    lambda row: gerar_mensagem(row["Nome"], row["Interesses"]), axis=1
)

print("Dados transformados:")
print(df.head())
```

Passo 4 – Carregamento (salvar os dados enriquecidos)

```python
# Salvar em novo CSV
novo_csv = "/content/usuarios_enriquecidos.csv"
df.to_csv(novo_csv, index=False)

# Salvar também em JSON
novo_json = "/content/usuarios_enriquecidos.json"
df.to_json(novo_json, orient="records", indent=4)

print(f"Arquivos gerados:\n- {novo_csv}\n- {novo_json}")
```
